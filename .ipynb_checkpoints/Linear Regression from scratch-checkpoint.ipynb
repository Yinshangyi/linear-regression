```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
def compute_error(a, b, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i,0]
        y = points[i,1]
        totalError += (y - (a*x + b))**2
    return totalError / float(len(points))
```

```python
def step_gradient(a_current, b_current, points, learningRate):
    # Initialize values of the gradients (according to B and M)
    a_gradient = 0
    b_gradient = 0
    
    N = float(len(points))
    
    for i in range(0, len(points)):
        x = points[i,0]
        y = points[i,1]
        a_gradient += -(2/N) * x * (y - ((a_current * x) + b_current))
        b_gradient += -(2/N) * (y - ((a_current * x) + b_current))
                
    new_b = b_current - (learningRate * b_gradient)
    new_a = a_current - (learningRate * a_gradient)
    return [new_a, new_b]
```

```python
def train(points, initial_a, initial_b, learning_rate, num_iteration):
    a = initial_a
    b = initial_b
    
    # List containing the values of a, b and error for each iteration
    values = []
    
    for i in range(num_iteration):
        a, b = step_gradient(a, b, points, learning_rate)
        values.append({'a' : a,
                       'b' : b,
                       'error' : compute_error(a,b,points)})
    return [a, b, values]
```

```python
def run():
    
    # Data collection
    points = genfromtxt('data.csv', delimiter=',')
    
    # Define hyperparameters
    learning_rate = 0.0001
    # y = mx + b
    # Initialization of A and B to 0
    initial_b = 0
    initial_a = 0
    num_iteration = 60000
        
    # Train our model
    print('Initially a = {0} and b = {1}'.format(initial_a, initial_b))
    print('Error (MSE) = {0}'.format(compute_error(initial_a, initial_b, points)))
    
    [a, b, values] = train(points, 
                   initial_a, 
                   initial_b,
                   learning_rate,
                   num_iteration)
    
    print('Training done')
    print('a = {0} and b = {1}'.format(a, b))
    print('Error (MSE) = {0}'.format(compute_error(a, b, points)))
    
    return [a,b,values]
```

```python
a,b,values = run()
```
